```python
#Generate random x values then generate y values using f(x)=1.3x-50, but give it a 10% chance of being a random value, this will create anomlies
import random
import pandas as pd
import numpy as np

X = [random.randrange(-1000,1000) for _ in range(0,1000)]
y = []

for x in X:
  prob = random.random()
  if prob<=.25: y.append(random.randrange(-1000,1000))
  else: y.append(x*1.3-50)

data = {"X":X,"y":y}
df = pd.DataFrame(data)
df.plot(x="X",y="y",kind="scatter",figsize=(13,5))
```

```python
def grad_decent(r,y,X,epochs,m=0,b=0):
  n = len(y)
  for _ in range(epochs):
    for i in range(n):
      #Calculate gradeint at given point m,b
      dm = (-2/n)*(y[i]-m*(X[i])-b)*X[i]
      db = (-1/n)*(y[i]-m*(X[i])-b)

      m = m - r*dm
      b = b - r*db

  return m,b
```

```python
m,b = grad_decent(.001,y,X,1000)
print(m,b)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue', label='Data points')
y_hat = [x*m+b for x in X]
plt.plot(X,y_hat)
# Plot the regression line
plt.show()
```

```python
def r_sqr(y,y_hat):
  y=np.array(y)
  y_hat = np.array(y_hat)
  y_bar = np.mean(y)

  tss = np.sum((y-y_hat)**2)
  rss = np.sum((y-y_bar)**2)

  return (1-tss/rss)

print(r_sqr(y,y_hat))
```
